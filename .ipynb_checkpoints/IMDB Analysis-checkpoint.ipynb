## Loading the necessary modules

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
import numpy as np
import plotly 
import re

from datetime import date
import matplotlib
matplotlib.style.use('ggplot')

%matplotlib inline
```

## Loading the dataset

Skipping the first 28 rows of the Excel file since they don't hold any relevant information.

```python
df = pd.read_excel("Programmes_MAL_v1.xls",sheetname="Sheet",skiprows=28,parse_cols=28)
```

```python
#df.head()
```

### Gender 

Define a gender function that returns the gender of a column. This will be used to make a male and female subsets.

```python
def gender(x):
    
    if re.search(r'^MY\s[W](.*?)',x):
        return "F"
    elif re.search(r'^MY\s[M](.*?)',x):
        return "M"
    elif x == "Channel Group ..." or x=="Date":
        return True
```

```python
gender_list = list(map(gender,df.columns))
```

Defining the Financial year `FY2016` and Month June for later use.

```python
FY2016 = df[df['Date'] >= date(2015,6,30)]
```

```python
FY2016 = FY2016[FY2016['Date'] <= date(2016,6,29)]
```

```python
JUNE16 = FY2016[FY2016['Date'] >=date(2016,6,1)]
```

At this point, I subset them based on gender.

```python
JUNE16_F = JUNE16.ix[:,[i for i in range(len(gender_list)) if ((gender_list[i] == "F") or (gender_list[i] == True))]] 
JUNE16_M = JUNE16.ix[:,[i for i in range(len(gender_list)) if ((gender_list[i] == "M") or (gender_list[i] == True))]] 

FY16_M = FY2016.ix[:,[i for i in range(len(gender_list)) if ((gender_list[i] == "M") or (gender_list[i] == True))]] 
FY16_F = FY2016.ix[:,[i for i in range(len(gender_list)) if ((gender_list[i] == "F") or (gender_list[i] == True))]] 


JUNE16_F_Agg = JUNE16_F.copy().groupby("Channel Group ...").sum().reset_index()
JUNE16_M_Agg = JUNE16_M.copy().groupby("Channel Group ...").sum().reset_index()
FY16_F_Agg = FY16_F.copy().groupby("Channel Group ...").sum().reset_index()
FY16_M_Agg = FY16_M.copy().groupby("Channel Group ...").sum().reset_index()
```

### Plots

```python
## necessary variables
ind = np.arange(4)                # the x locations for the groups
width = 0.35


plt.figure()


ax1 =JUNE16_M_Agg.plot(kind= 'bar')
ax1.set_ylabel('Sum')
ax1.set_title('Male Viewer age demographic in June 2016')
xTickMarks = JUNE16_F_Agg["Channel Group ..."].unique()
ax1.set_xticks(ind)
xtickNames = ax1.set_xticklabels(xTickMarks)
plt.setp(xtickNames, rotation=60, fontsize=12)

ax =JUNE16_F_Agg.plot(kind= 'bar')
ax.set_ylabel('Sum')
ax.set_title('Female Viewer age demographic in June 2016')
xTickMarks = JUNE16_F_Agg["Channel Group ..."].unique()
ax.set_xticks(ind)
xtickNames = ax.set_xticklabels(xTickMarks)
plt.setp(xtickNames, rotation=60, fontsize=12)



ax2 =FY16_M_Agg.plot(kind= 'bar')
ax2.set_ylabel('Sum')
ax2.set_title('Male Viewer age demographic in Financial Year 2016')
xTickMarks = JUNE16_F_Agg["Channel Group ..."].unique()
ax2.set_xticks(ind)
xtickNames = ax2.set_xticklabels(xTickMarks)
plt.setp(xtickNames, rotation=60, fontsize=12)


ax3 =FY16_F_Agg.plot(kind= 'bar')
ax3.set_ylabel('Sum')
ax3.set_title('Female Viewer age demographic in Financial Year 2016')
xTickMarks = JUNE16_F_Agg["Channel Group ..."].unique()
ax3.set_xticks(ind)
xtickNames = ax3.set_xticklabels(xTickMarks)
plt.setp(xtickNames, rotation=60, fontsize=12)

plt.show()
```

## Racial Demographics

```python
def race(x):
    if re.search(r'^MY\s\w*\s\w*',x):
        if re.search(r'female$',x):
            return "F"
        else:
            return "M"
    elif re.search(r'^Channel(.*?)',x):
        return True
    else:
        return False
```

```python
race_gender_list = list(map(race,df.columns))
```

```python
FY16_R_M = FY2016.ix[:,[i for i in range(len(race_gender_list)) if ((race_gender_list[i] == "M")
                                                                    or (race_gender_list[i] == True))]] 
FY16_R_F = FY2016.ix[:,[i for i in range(len(race_gender_list)) if ((race_gender_list[i] == "F")
                                                                    or (race_gender_list[i] == True))]] 

JUNE16_R_M = JUNE16.ix[:,[i for i in range(len(race_gender_list)) if ((race_gender_list[i] == "M")
                                                                    or (race_gender_list[i] == True))]] 
JUNE16_R_F = JUNE16.ix[:,[i for i in range(len(race_gender_list)) if ((race_gender_list[i] == "F")
                                                                    or (race_gender_list[i] == True))]] 
```

```python
JUNE16_R_F_Agg = JUNE16_R_F.copy().groupby("Channel Group ...").sum().reset_index()
JUNE16_R_M_Agg = JUNE16_R_M.copy().groupby("Channel Group ...").sum().reset_index()
FY16_R_F_Agg = FY16_R_F.copy().groupby("Channel Group ...").sum().reset_index()
FY16_R_M_Agg = FY16_R_M.copy().groupby("Channel Group ...").sum().reset_index()
```

### PLOTS

```python
## necessary variables
ind = np.arange(4)                # the x locations for the groups
width = 0.35


plt.figure()


ax1 =JUNE16_R_M_Agg.plot(kind= 'bar')
ax1.set_ylabel('Sum')
ax1.set_title('Male Viewer racial demographic in June 2016')
xTickMarks = JUNE16_R_F_Agg["Channel Group ..."].unique()
ax1.set_xticks(ind)
xtickNames = ax1.set_xticklabels(xTickMarks)
plt.setp(xtickNames, rotation=60, fontsize=12)

ax =JUNE16_R_F_Agg.plot(kind= 'bar')
ax.set_ylabel('Sum')
ax.set_title('Female Viewer racial demographic in June 2016')
xTickMarks = JUNE16_R_F_Agg["Channel Group ..."].unique()
ax.set_xticks(ind)
xtickNames = ax.set_xticklabels(xTickMarks)
plt.setp(xtickNames, rotation=60, fontsize=12)



ax2 =FY16_R_M_Agg.plot(kind= 'bar')
ax2.set_ylabel('Sum')
ax2.set_title('Male Viewer racial demographic in Financial Year 2016')
xTickMarks = JUNE16_R_F_Agg["Channel Group ..."].unique()
ax2.set_xticks(ind)
xtickNames = ax2.set_xticklabels(xTickMarks)
plt.setp(xtickNames, rotation=60, fontsize=12)


ax3 =FY16_R_F_Agg.plot(kind= 'bar')
ax3.set_ylabel('Sum')
ax3.set_title('Female Viewer racial demographic in Financial Year 2016')
xTickMarks = JUNE16_R_F_Agg["Channel Group ..."].unique()
ax3.set_xticks(ind)
xtickNames = ax3.set_xticklabels(xTickMarks)
plt.setp(xtickNames, rotation=60, fontsize=12)

plt.show()
```
